```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('graB.csv')
```

```python
df
```

```python
target=[]
for row in df['ΔΔG (Mt WT)']:
 if row<0 : target.append(0)
 else: target.append(1);
```

```python
df['target'] = target
df
```

```python
df.drop(['WT/Mutant'], axis='columns', inplace=True)
df.head(5)
```

```python
target = df.target
train_df = df.drop('target', axis='columns')
```

```python
train_df.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_df, target, test_size=0.2, random_state=0)
```

```python
len(x_train)
```

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
NB.fit(x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(5)
knn.fit(x_train, y_train)
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(x_train, y_train)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
len(x_train.columns)
```

```python
ANN = keras.Sequential([
    keras.layers.Dense(10, input_shape =(10,), activation='relu'),
    keras.layers.Dense(1,  activation='sigmoid'),
    
])
ANN.compile(
        optimizer = 'adam',
         loss = 'binary_crossentropy',
           metrics = ['accuracy'])
ANN.fit(x_train, y_train, epochs =200)
```

```python
ANN.summary()
```

```python
ANN.evaluate(x_test, y_test)
```

```python
from sklearn.metrics import plot_roc_curve
```

```python
disp = plot_roc_curve(NB, x_test, y_test)
plot_roc_curve(knn, x_test, y_test, ax=disp.ax_);
plot_roc_curve(svm, x_test, y_test, ax=disp.ax_);
```

```python
from sklearn.model_selection import cross_val_score
```

```python
svm_cv_Score = np.mean(cross_val_score(SVC(), train_df, target, cv = 10))*100
svm_cv_Score
```

```python
knn_cv_Score = np.mean(cross_val_score(KNeighborsClassifier(5), train_df, target, cv = 10))*100
knn_cv_Score
```

```python
NB_cv_Score = np.mean(cross_val_score(GaussianNB(), train_df, target, cv = 10))*100
NB_cv_Score
```
